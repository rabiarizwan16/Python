```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#np.seed
data={
    "Date":pd.date_range(start="2024-01-01", periods=100, freq="D"),
    "Product":np.random.choice(["TATA Salt","Buiscuit","Kit Kat"],100),
    "Region":np.random.choice(["North","South","East","West"],100),
    "Units Sold":np.random.randint(1,50,size=100),
    "Revenue":np.random.randint(50,500, size=100)
}
sales_data=pd.DataFrame(data)
print(sales_data.head())
    
```

```python
data={
    "Date":pd.date_range(start="2024-01-01", periods=100, freq="D"),
    "Product":np.random.choice(["TATA Salt","Buiscuit","Kit Kat"],100),
    "Region":np.random.choice(["North","South","East","West"],100),
    "Units Sold":np.random.randint(1,50,size=100),
    "Revenue":np.random.randint(50,500, size=100)
}
sales_data=pd.DataFrame(data)
print(sales_data.head())
```

```python
total_revenue=sales_data["Revenue"].sum()
print(total_revenue)
```

```python
total_units_sold=sales_data["Units Sold"].sum()
print(total_units_sold)
```

```python
avg_revenue_per_sale=sales_data["Revenue"].mean()
print(avg_revenue_per_sale)
```

```python

fig,axs =plt.subplots(3,1, figsize=(12,15), constrained_layout=True)
region_revenue=sales_data.groupby("Region")["Revenue"].sum()
axs[0].bar(region_revenue.index,region_revenue.values,color='steelblue',edgecolor='black')
axs[0].set_title('Total Revenue by Region',fontsize=12)
axs[0].set_xlabel('Total Revenue by Region',fontsize=12)
axs[0].set_ylabel('Total Revenue by Region',fontsize=12)
plt.show()
```
